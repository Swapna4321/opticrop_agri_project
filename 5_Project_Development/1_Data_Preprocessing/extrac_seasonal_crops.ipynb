```python
import pandas as pd

# Load Dataset
df = pd.read_csv("Crop_recommendation.csv")

# Sample Seasonal Classification
def get_season(rainfall):
    if rainfall < 100:
        return "Summer"
    elif rainfall < 200:
        return "Winter"
    else:
        return "Monsoon"

df['Season'] = df['rainfall'].apply(get_season)

print(df[['label', 'Season']].head())

# Save Updated Dataset
df.to_csv("seasonal_crop_data.csv", index=False)
```
